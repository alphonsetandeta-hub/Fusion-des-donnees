#### Détection et Correction des Valeurs Aberrantes (Outliers)

Pour une détection plus systématique, nous utiliserons la méthode de l'IQR (Interquartile Range). Une valeur est considérée comme aberrante si elle est inférieure à `Q1 - 1.5 * IQR` ou supérieure à `Q3 + 1.5 * IQR`.

```python
# Détecter les valeurs aberrantes (outliers) en utilisant la méthode IQR sur df_cleaned (incluant ETP)

outliers_count = {}
# df_cleaned = df.copy() # Cette ligne est commentée car df_cleaned est déjà la version nettoyée et mise à jour

for col in numeric_cols_updated:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identifier les outliers
    col_outliers = df_cleaned[(df_cleaned[col] < lower_bound) | (df_cleaned[col] > upper_bound)]
    num_outliers = len(col_outliers)

    if num_outliers > 0:
        outliers_count[col] = num_outliers

print("Nombre de valeurs aberrantes (outliers) détectées par colonne dans 'df_cleaned' (méthode IQR) :")
if outliers_count:
    for col, count in outliers_count.items():
        print(f"- {col}: {count}")
else:
    print("Aucune valeur aberrante significative détectée dans les colonnes numériques de 'df_cleaned'.")
```

### Détection des valeurs aberrantes par la méthode IQR

Cette section implémente la méthode de l'intervalle interquartile (IQR) pour détecter les valeurs aberrantes dans toutes les colonnes numériques du DataFrame `df_cleaned`. La méthode IQR identifie les valeurs qui sont significativement éloignées des autres données, au-delà de 1.5 fois l'IQR des premier et troisième quartiles. Le code vérifie et affiche le nombre d'outliers pour chaque colonne, préparant ainsi le terrain pour leur correction.

#### Stratégie de Correction des Outliers : Le Plafonnement (Capping)

Pour corriger ces valeurs aberrantes, une méthode courante est le **plafonnement (capping)**, qui consiste à remplacer les valeurs aberrantes situées en dessous de la limite inférieure par la limite inférieure, et celles situées au-dessus de la limite supérieure par la limite supérieure. Cette méthode permet de conserver toutes les observations tout en réduisant l'influence des valeurs extrêmes. Nous appliquerons cette correction sur une copie du DataFrame pour ne pas modifier les données brutes.

```python
# Appliquer la correction des outliers par plafonnement (capping) sur df_cleaned (incluant ETP)
# df_cleaned est déjà la version nettoyée et mise à jour avec l'ETP.

for col in numeric_cols_updated:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Plafonner les valeurs aberrantes
    df_cleaned[col] = np.where(df_cleaned[col] < lower_bound, lower_bound, df_cleaned[col])
    df_cleaned[col] = np.where(df_cleaned[col] > upper_bound, upper_bound, df_cleaned[col])

print("Les valeurs aberrantes (y compris pour l'ETP) ont été corrigées par plafonnement dans le DataFrame 'df_cleaned'.")

# Vérifier les statistiques descriptives après correction
print("Statistiques descriptives après correction des outliers (mise à jour) :")
display(df_cleaned[numeric_cols_updated].describe())
```

#### Calcul de l'évapotranspiration potentielle (ETP) - Méthode FAO Penman-Monteith

Nous allons utiliser la méthode standardisée FAO-56 Penman-Monteith pour calculer l'évapotranspiration de référence ($ET_0$). Cette méthode est recommandée car elle fournit des estimations cohérentes de l'ETP en utilisant les données météorologiques disponibles.

La formule de Penman-Monteith est la suivante :

$ET_0 = \frac{0.408 \Delta (R_n - G) + \gamma \frac{900}{T_{mean} + 273} u_2 (e_s - e_a)}{\Delta + \gamma (1 + 0.34 u_2)}$

Où :
- $ET_0$ : Évapotranspiration de référence (mm/jour)
- $\Delta$ : Pente de la courbe de pression de vapeur saturante (kPa/°C)
- $R_n$ : Rayonnement net à la surface de la culture (MJ m$^{-2}$ jour$^{-1}$)
- $G$ : Flux de chaleur du sol (MJ m$^{-2}$ jour$^{-1}$) - généralement négligé pour les calculs journaliers ($G \approx 0$)
- $\gamma$ : Constante psychrométrique (kPa/°C)
- $T_{mean}$ : Température moyenne de l'air (°C)
- $u_2$ : Vitesse du vent à 2 m de hauteur (m/s)
- $e_s$ : Pression de vapeur saturante moyenne (kPa)
- $e_a$ : Pression de vapeur réelle (kPa)
- $900$ : Constante de conversion
- $0.34$ : Constante de la résistance aérodynamique

Nous devrons calculer plusieurs variables intermédiaires pour arriver à $ET_0$.

### Calcul de l'évapotranspiration potentielle (ETP) - Méthode FAO Penman-Monteith

Cette section est dédiée à l'ingénierie de caractéristique la plus complexe : le calcul de l'évapotranspiration de référence (ET0) à l'aide de la méthode FAO Penman-Monteith. Cette méthode est standardisée et utilise plusieurs variables climatiques (`Tmin`, `Tmax`, humidité, vitesse du vent, rayonnement solaire) pour estimer l'évapotranspiration d'une culture de référence. Le code effectue une série de calculs intermédiaires (pression de vapeur saturante, rayonnement extraterrestre, rayonnement net, etc.) pour aboutir à la valeur finale de ET0, qui est ensuite ajoutée au DataFrame `df_cleaned`. ET0 est une variable agronomique cruciale pour comprendre les besoins en eau de la culture et son impact sur la biomasse.

```python
import numpy as np

# --- 1. Constantes et Paramètres de base ---
latitude_deg = 14.7  # Latitude approximative du Sénégal (ex: Dakar)
latitude_rad = np.radians(latitude_deg) # Convertir en radians
Gsc = 0.0820  # Constante solaire (MJ m^-2 min^-1)
sigma = 4.903e-9 # Constante de Stefan-Boltzmann (MJ m^-2 K^-4 jour^-1)
albedo = 0.23 # Albedo pour une culture de référence (gazon)
P = 101.3 # Pression atmosphérique standard au niveau de la mer (kPa)

# --- 2. Fonction pour la pression de vapeur saturante ---
def eo_T(T):
    return 0.6108 * np.exp(17.27 * T / (T + 237.3))

# --- 3. Rayonnement extraterrestre (Ra) ---
def calculate_ra(day_of_year, latitude_rad, Gsc_val):
    dr = 1 + 0.033 * np.cos(2 * np.pi * day_of_year / 365)
    delta = 0.409 * np.sin(2 * np.pi * day_of_year / 365 - 1.39)

    tan_prod = -np.tan(latitude_rad) * np.tan(delta)
    tan_prod = np.clip(tan_prod, -0.99999, 0.99999) # Clamper les valeurs pour arccos
    ws = np.arccos(tan_prod)

    Ra = (24 * 60 / np.pi) * Gsc_val * dr * \
         (ws * np.sin(latitude_rad) * np.sin(delta) + \
          np.cos(latitude_rad) * np.cos(delta) * np.sin(ws))
    return Ra

# Créer une copie pour les calculs d'ETP afin de ne pas modifier l'original df_cleaned pré-outliers
df_et0 = df_cleaned.copy()

# Jour de l'année
df_et0['day_of_year'] = df_et0['date'].dt.dayofyear

# Calcul de Ra
df_et0['Ra'] = df_et0.apply(lambda row: calculate_ra(row['day_of_year'], latitude_rad, Gsc), axis=1)

# --- 4. Calcul des composants Penman-Monteith ---

# Températures en Kelvin
df_et0['T_min_K'] = df_et0['temp_min_c'] + 273.15
df_et0['T_max_K'] = df_et0['temp_max_c'] + 273.15

# Pression de vapeur saturante (es)
df_et0['es_min'] = eo_T(df_et0['temp_min_c'])
df_et0['es_max'] = eo_T(df_et0['temp_max_c'])
df_et0['es'] = (df_et0['es_min'] + df_et0['es_max']) / 2

# Pression de vapeur réelle (ea) - en utilisant Humidité Relative min et max
df_et0['ea'] = (df_et0['es_min'] * (df_et0['hum_rel_max_pct'] / 100) + \
                df_et0['es_max'] * (df_et0['hum_rel_min_pct'] / 100)) / 2

# Pente de la courbe de pression de vapeur saturante (Delta)
df_et0['Delta'] = (4098 * eo_T(df_et0['temp_moy_c'])) / (df_et0['temp_moy_c'] + 237.3)**2

# Constante psychrométrique (gamma)
df_et0['gamma'] = 0.000665 * P

# Rayonnement net à ondes courtes (Rns)
df_et0['Rns'] = (1 - albedo) * df_et0['rad_sol_mjm2']

# Rayonnement clair (Rso) - Assumant Z=0 pour le moment
df_et0['Rso'] = 0.75 * df_et0['Ra']

# Calcul du ratio Rs/Rso, plafonné à 1.0
df_et0['Rs_Rso'] = np.minimum(df_et0['rad_sol_mjm2'] / df_et0['Rso'], 1.0)

# Rayonnement net à ondes longues (Rnl)
Rnl_factor1 = sigma * ((df_et0['T_max_K']**4 + df_et0['T_min_K']**4) / 2)
Rnl_factor2 = (0.34 - 0.14 * np.sqrt(df_et0['ea']))
Rnl_factor3 = (1.35 * df_et0['Rs_Rso'] - 0.35)

df_et0['Rnl'] = Rnl_factor1 * Rnl_factor2 * Rnl_factor3

# Rayonnement net (Rn)
df_et0['Rn'] = df_et0['Rns'] - df_et0['Rnl']

# Flux de chaleur du sol (G) - Assumé 0 pour les calculs journaliers
df_et0['G'] = 0

# --- 5. Calcul final de ET0 Penman-Monteith ---

ET0_Numerator = (0.408 * df_et0['Delta'] * (df_et0['Rn'] - df_et0['G'])) + \
                (df_et0['gamma'] * (900 / (df_et0['temp_moy_c'] + 273.15)) * \
                 df_et0['vent_ms'] * (df_et0['es'] - df_et0['ea']))

ET0_Denominator = df_et0['Delta'] + df_et0['gamma'] * (1 + 0.34 * df_et0['vent_ms'])

df_et0['ET0_Penman_Monteith'] = ET0_Numerator / ET0_Denominator

# Ajouter la colonne ET0_Penman_Monteith au DataFrame original nettoyé
df_cleaned['ET0_Penman_Monteith'] = df_et0['ET0_Penman_Monteith']

print("Colonne 'ET0_Penman_Monteith' ajoutée au DataFrame 'df_cleaned'.")

# Afficher les premières lignes avec la nouvelle colonne
display(df_cleaned[['date', 'temp_min_c', 'temp_max_c', 'temp_moy_c', 'hum_rel_min_pct', 'hum_rel_max_pct', 'vent_ms', 'rad_sol_mjm2', 'ET0_Penman_Monteith']].head())
```

La colonne `ET0_Penman_Monteith` a été calculée et ajoutée à votre DataFrame `df_cleaned`. Cette colonne représente l'évapotranspiration de référence quotidienne en mm/jour, estimée selon la méthode FAO Penman-Monteith.

### Distribution des Variables Numériques avec Densité, Moyenne et Coefficient de Variation

### Distributions des variables numériques avec densité, moyenne et Coefficient de Variation (CV)

Cette section génère des histogrammes avec des courbes de densité (KDE) pour toutes les variables numériques du DataFrame. Pour chaque graphique, la moyenne est indiquée par une ligne verticale et le Coefficient de Variation (CV) est affiché. Le CV (écart-type / moyenne) est une mesure de la variabilité relative, utile pour comparer la dispersion de variables ayant des échelles différentes. Ces visualisations permettent d'évaluer la forme des distributions (normale, asymétrique, etc.) et la dispersion relative de chaque variable.

```python
from scipy.stats import variation

# Get all numerical columns from df_cleaned
numeric_cols = df_cleaned.select_dtypes(include=np.number).columns

# Determine the number of rows and columns for the subplot grid
num_plots = len(numeric_cols)
num_cols_subplot = 3  # You can adjust this for desired layout
num_rows_subplot = (num_plots + num_cols_subplot - 1) // num_cols_subplot

plt.figure(figsize=(num_cols_subplot * 6, num_rows_subplot * 5))

for i, col in enumerate(numeric_cols):
    plt.subplot(num_rows_subplot, num_cols_subplot, i + 1)

    # Plot histogram and KDE
    sns.histplot(df_cleaned[col], kde=True, bins=30)
    plt.title(f'Distribution de {col}')
    plt.xlabel(col)
    plt.ylabel('Fréquence / Densité')

    # Calculate mean and Coefficient of Variation (CV)
    mean_val = df_cleaned[col].mean()
    std_val = df_cleaned[col].std()
    # Avoid division by zero if mean is 0
    cv_val = (std_val / mean_val) if mean_val != 0 else np.nan

    # Add mean and CV to the plot
    plt.axvline(mean_val, color='r', linestyle='--', label=f'Moyenne: {mean_val:.2f}')
    plt.text(x=mean_val + (plt.xlim()[1] - plt.xlim()[0]) * 0.05, y=plt.ylim()[1] * 0.9,
             s=f'Moyenne: {mean_val:.2f}\nCV: {cv_val:.2f}',
             color='red', bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))

    plt.legend()

plt.tight_layout()
plt.show()
```

### Affichage des distributions de toutes les variables numériques

Cette série de graphiques présente la distribution de chaque variable numérique du DataFrame. Chaque histogramme est accompagné d'une estimation de densité (KDE), d'une ligne indiquant la moyenne et d'un texte affichant la moyenne et le Coefficient de Variation (CV). Ces visualisations sont fondamentales pour comprendre les caractéristiques intrinsèques de chaque variable, leur étendue, leur symétrie et leur variabilité, ce qui est essentiel pour la modélisation future.

### Distribution des 10 Variables Numériques avec le CV le plus élevé

### Focus sur les 10 variables numériques avec le CV le plus élevé

Pour une analyse plus ciblée de la variabilité, cette section identifie et visualise les distributions des 10 variables numériques qui présentent le Coefficient de Variation (CV) le plus élevé. Un CV élevé indique une grande dispersion relative des données par rapport à leur moyenne, suggérant que ces variables pourraient être très discriminantes ou potentiellement plus influentes dans les modèles. L'analyse de ces distributions permet de mieux comprendre les sources de variabilité les plus importantes dans le jeu de données.

```python
from scipy.stats import variation

# Recalculer les CV pour toutes les colonnes numériques
numeric_cols = df_cleaned.select_dtypes(include=np.number).columns
cv_values = {}
for col in numeric_cols:
    mean_val = df_cleaned[col].mean()
    std_val = df_cleaned[col].std()
    if mean_val != 0:
        cv_values[col] = std_val / mean_val
    else:
        cv_values[col] = np.nan # Gérer le cas où la moyenne est zéro

# Convertir en Series pour faciliter le tri
cv_series = pd.Series(cv_values).dropna()

# Sélectionner les 10 premières colonnes avec le CV le plus élevé
top_10_cv_cols = cv_series.nlargest(10).index.tolist()

print("Top 10 des variables numériques avec le CV le plus élevé:")
print(top_10_cv_cols)

# Générer les plots pour ces 10 variables
num_plots = len(top_10_cv_cols)
num_cols_subplot = 3
num_rows_subplot = (num_plots + num_cols_subplot - 1) // num_cols_subplot

plt.figure(figsize=(num_cols_subplot * 6, num_rows_subplot * 5))

for i, col in enumerate(top_10_cv_cols):
    plt.subplot(num_rows_subplot, num_cols_subplot, i + 1)

    # Plot histogram and KDE
    sns.histplot(df_cleaned[col], kde=True, bins=30)
    plt.title(f'Distribution de {col}')
    plt.xlabel(col)
    plt.ylabel('Fréquence / Densité')

    # Calculate mean and CV (already calculated, just retrieve)
    mean_val = df_cleaned[col].mean()
    cv_val = cv_series[col] # Utiliser la valeur calculée

    # Add mean and CV to the plot
    plt.axvline(mean_val, color='r', linestyle='--', label=f'Moyenne: {mean_val:.2f}')
    plt.text(x=mean_val + (plt.xlim()[1] - plt.xlim()[0]) * 0.05, y=plt.ylim()[1] * 0.9,
             s=f'Moyenne: {mean_val:.2f}\nCV: {cv_val:.2f}',
             color='red', bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))

    plt.legend()

plt.tight_layout()
plt.show()
```

### Affichage des distributions des top 10 CV variables

```python
from collections import defaultdict

# Redéfinir les groupes de variables avec les noms harmonisés (pour être sûr)
phenological_cols_list = ['temp', 'humidity', 'ec', 'ph', 'nitrogen', 'phosphorus', 'potassium', 'salinity']
indice_cols_list = ['ndvi', 'ndre', 'savi', 'cire', 'ndwi', 'lai', 'fapar']
climatiques_cols_list = ['Tmin', 'Tavg', 'Tmax', 'RHmin', 'RHavg', 'RHmax', 'WindSpeed', 'SolarRad_Jcm2', 'SolarRad_Mjm2'] # Updated list
radar_cols_list = ['vv', 'vh', 'rvi', 'ssm']
additional_metrics_cols_list = ['das', 'ET0', 'biomasse_cum']

# Créer un mapping de chaque colonne à sa catégorie
column_to_category = {}
for col in phenological_cols_list: column_to_category[col] = 'Phénologiques'
for col in indice_cols_list: column_to_category[col] = 'Indice'
for col in climatiques_cols_list: column_to_category[col] = 'Climatiques'
for col in radar_cols_list: column_to_category[col] = 'Radar'
for col in additional_metrics_cols_list: column_to_category[col] = 'Métriques Additionnelles'

# Répartir les top 10 CV colonnes par catégorie
top_10_cv_by_category = defaultdict(list)
for col in top_10_cv_cols:
    category = column_to_category.get(col, 'Autre') # Gérer les colonnes non catégorisées si besoin
    top_10_cv_by_category[category].append(col)

# Générer les plots pour chaque catégorie
for category, cols_to_plot in top_10_cv_by_category.items():
    if not cols_to_plot:
        continue

    print(f"\n### Distributions des variables '{category}' (Top 10 CV) ###")

    num_plots = len(cols_to_plot)
    num_cols_subplot = 3
    num_rows_subplot = (num_plots + num_cols_subplot - 1) // num_cols_subplot

    fig = plt.figure(figsize=(num_cols_subplot * 6, num_rows_subplot * 5))

    for i, col in enumerate(cols_to_plot):
        plt.subplot(num_rows_subplot, num_cols_subplot, i + 1)

        # Plot histogram and KDE
        sns.histplot(df_cleaned[col], kde=True, bins=30)
        plt.title(f'Distribution de {col}')
        plt.xlabel(col)
        plt.ylabel('Fréquence / Densité')

        # Retrieve mean and CV
        mean_val = df_cleaned[col].mean()
        cv_val = cv_series[col] # Utiliser la valeur calculée

        # Add mean and CV to the plot
        plt.axvline(mean_val, color='r', linestyle='--', label=f'Moyenne: {mean_val:.2f}')
        plt.text(x=mean_val + (plt.xlim()[1] - plt.xlim()[0]) * 0.05, y=plt.ylim()[1] * 0.9,
                 s=f'Moyenne: {mean_val:.2f}\nCV: {cv_val:.2f}',
                 color='red', bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))

        plt.legend()

    plt.tight_layout()

    # Export the plot
    output_filename = f'distribution_top10cv_{category}.png'
    plt.savefig(output_filename)
    print(f"Le graphique de distribution pour les variables '{category}' a été exporté vers '{output_filename}'.")
    plt.show()
```

```python
# Visualisation de la biomasse cumulative
plt.figure(figsize=(15, 6))
sns.lineplot(x='date', y='biomasse_cum', data=df_cleaned_sorted, color='purple')
plt.title('Évolution Temporelle de la Biomasse Cumulative avec Stades Phénologiques')
plt.xlabel('Date')
plt.ylabel('Biomasse Cumulative')
plt.grid(True, linestyle='--', alpha=0.7)

# Définir les stades phénologiques et leurs plages de DAS
phenological_stages = {
    'Levée': (0, 10),
    'Tallage': (10, 25),
    'Montaison': (25, 45),
    'Épiaison': (45, 65),
    'Floraison': (65, 75),
    'Remplissage': (75, 90),
    'Maturité': (90, 110)
}

# Trouver la date de début pour ancrer les DAS
min_date = df_cleaned_sorted['date'].min()

# Ajouter les stades phénologiques aux graphiques
for stage, (das_start, das_end) in phenological_stages.items():
    start_date = min_date + pd.to_timedelta(das_start, unit='D')
    end_date = min_date + pd.to_timedelta(das_end, unit='D')

    plt.axvspan(start_date, end_date, color='gray', alpha=0.2, label=stage)
    mid_date = start_date + (end_date - start_date) / 2
    plt.text(mid_date, plt.ylim()[1] * 0.95, stage, rotation=90, va='top', ha='center', color='black', fontsize=9)

plt.tight_layout()
plt.show()
```
